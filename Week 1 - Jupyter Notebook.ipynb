```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Create Scatter Plot

```python
x = [3, 5, 6]
y = [4, 8, 3]

plt.scatter(x, y)
plt.show()
```

# Create Line Chart

```python
x = [3, 5, 6]
y = [4, 8, 3]

plt.plot(x, y)
plt.show()
```

# Create Scatterplot with Many Points

```python
x = []
y = []

for i in range(1000):
    x.append(i * np.random.rand())
    y.append(i * np.random.rand())
    

plt.scatter(x, y)
plt.show()
```
